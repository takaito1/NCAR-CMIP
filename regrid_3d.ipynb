# regridding of a selected 3D variable

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
import os
```

```python
# select a variable
var='po4'
df=pd.read_csv(var+'.txt')
dirs=df['dirs']
df
```

```python
# regrid
df=pd.read_csv(var+'.txt')
dirs=df['dirs']
N=len(dirs)
os.system('mkdir -p bil_woagrid')
for n in range(4,N):
    model=dirs[n].split('/')[7]
    os.system('ls '+model+'/'+var+'*.nc > tmp.txt')
    fid=open('tmp.txt','r')
    files=fid.read().splitlines()
    for m in files:
        os.system('cdo intlevel,0,10,20,30,50,75,100,125,150,200,250,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1750,2000,2500,3000,3500,4000,4500,5000,5500 '+m+' temp.nc')
        os.system('mkdir -p bil_woagrid/'+model)
        os.system('cdo remapbil,grd1x1 temp.nc bil_woagrid/'+m+'_woagrid.nc')
```

```python
# merge gn files
df=pd.read_csv(var+'.txt')
dirs=df['dirs']
N=len(dirs)
for n in range(0,N):
    model=dirs[n].split('/')[7]
    os.system('mkdir -p woagrid/'+model)
    os.system('cdo mergetime bil_woagrid/'+model+'/'+var+'*gn*_woagrid.nc woagrid/'+model+'/'+var+'_'+model+'_Omon_historical_r1i1p1f1_gn_woagrid.nc')
```

```python
# merge gr files
df=pd.read_csv(var+'.txt')
dirs=df['dirs']
N=len(dirs)
for n in range(0,N):
    model=dirs[n].split('/')[7]
    os.system('mkdir -p woagrid/'+model)
    os.system('cdo mergetime bil_woagrid/'+model+'/'+var+'*gr*_woagrid.nc woagrid/'+model+'/'+var+'_'+model+'_Omon_historical_r1i1p1f1_gr_woagrid.nc')
```
